```python
import lcmap.db.tile as tile
from lcmap.ingest.util import snap, to_lon_lat, to_epsg_5070
from lcmap.db.util import as_array
from random import randint
import numpy as np
```

```python
# Boundaries for random area
lower_x, lower_y = to_epsg_5070(lon = -118, lat = 46.5)
upper_x, upper_y = to_epsg_5070(lon = -123, lat = 48.5)
layers = ["sr_band1", "sr_band2", "sr_band3", "sr_band4", "sr_band5", "sr_band6", "sr_band7"]
```

```python
def randomly_find_tile(start = '2010-01-01', stop = '2015-06-01'):
    # random point
    rx, ry = randint(upper_x, lower_x), randint(lower_y, upper_y)
    x,  y  = snap(tx, ty, tile_size=256, offset_y=-20)
    # random layer
    layer_idx = randint(0, 6)
    layer = layers[layer_idx]
    # random year
    rows = tile.find(x, y, layer, start, stop)
```

```python
obj = %timeit -n1 -r10 -o randomly_find_tile('2014-01-01','2014-02-01')
month_a = obj
"best: {0}ms, worst: {1}ms, total: {2}".format(obj.best*1000, obj.worst*1000, obj.all_runs)
```
