```python
from lcmap_client import Client
import numpy as np
import numpy.ma as ma
```

```python
client = Client();
```

```python
cloud = "LANDSAT_8/OLI_TIRS/cfmask"
red   = "LANDSAT_8/OLI_TIRS/toa_band4"
green = "LANDSAT_8/OLI_TIRS/toa_band3"
blue  = "LANDSAT_8/OLI_TIRS/toa_band2"
t1 = "2014-01-01"
t2 = "2015-01-01"
x = -1910865
y = 2956785
```

```python
_, red_tiles = client.data.surface_reflectance.tiles(red, x, y, t1, t2);
_, green_tiles = client.data.surface_reflectance.tiles(green, x, y, t1, t2);
_, blue_tiles = client.data.surface_reflectance.tiles(blue, x, y, t1, t2);
```

```python
r,g,b = red_tiles[0].data, green_tiles[0].data, blue_tiles[0].data;
```
